# DataViz Discrimination Excersise

A few years ago, a discrimination lawsuit was filed against the California DDS claiming that White Non-Hispanics were receiving more funding than Hispanics. If this was true, the DDS faced some serious implications. In order to investigate the issue, several statisticians were hired to further analyze the data.

After viewing the summary, this sample of 1000 individuals seems like a good representation of the state of California as a whole. The ages look to be semi-uniformly distributed, and the ethnicity frequencies seem to accurately represent the demographic proportions in the state.

Now, we are interested in determining if there is any ethnic bias regarding the expenditures.

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/KaonToPion/datasets/main/californiaDDSDataV2.csv'
df = pd.read_csv(url)
```

```python
df.shape
```

```python
df.head()
```

```python
# Cleaning the dataset from non White not Hispanic and Hispanic ethnic groups
df_clean = df.loc[(df['Ethnicity'] == 'White not Hispanic') | (df['Ethnicity'] == 'Hispanic')].copy()
df_clean.shape
```

```python
# Histogram of expenditures according to ethnicity
sns.displot(data=df_clean, x='Expenditures', hue='Ethnicity', legend=True, fill=True, kind='kde');
```

We clearly see a two differences:

- Hispanics receive more low quantity funds
- White not Hispanic receive more high quantity funds

This could be because there are more old-age white not hispanic than hispanic, old-age people receive more expenditures than other age cohorts.

It's necessary to carry out a more thorough analysis.

```python
fig, ax1 = plt.subplots(figsize=(12, 8))
sns.swarmplot(data=df_clean, x='Expenditures', y='Ethnicity', hue ='Age Cohort', ax=ax1);
```

```python
sns.barplot(data=df_clean, x='Age Cohort', y='Expenditures', hue='Ethnicity');
```

There is no difference between the fundings that receive each ethnic group.

# Visualización de datos con python

La visualización de datos es una representación gráfica de la información y los datos. Mediante el uso de elementos visuales, como gráficos y mapas, la visualización de datos ofrece una manera accesible de detectar y comprender tendencias, valores atípicos y los patrones en los datos. En el mundo del big data, las herramientas y las tecnologías de visualización de datos son esenciales para analizar cantidades masivas de información y tomar decisiones basadas en los datos.

También es una herramienta muy útil a la hora de transmitir información, y de facilitar la compresión de las partes más cuantitavas de un análisis.

Dentro del ecosistema de Python se pueden encontrar multitud de librerías que están desarrolladas específicamente para la visualización de datos, algunas de las que se verán en este notebook son: 

- `matplotlib`
- `seaborn`

### 1.2.1. Líneas

Graficar mediante una línea el valor de la función seno y coseno, en el intervalo $[-\pi, \pi]$:

````python
x = np.linspace(-np.pi, np.pi, 100)
cos, sin = np.cos(x), np.sin(x)

plt.plot(x, cos)
plt.plot(x, sin);
````

Graficar mediante una línea el valor de la función $f(x)$, en el intervalo $[-7, 7]$:

$$f(x) = \frac{e^x}{1 + e^x}$$

````python
x = np.linspace(-7, 7, 100)
y = np.exp(x) / (1 + np.exp(x))

plt.plot(x, y);
````

Mediante el método `plt.fill_between()`, se pueden establecer condiciones para colorear areas entre varias funciones.

````python
x = np.linspace(-5, 5, 100)
y1 = np.cos(x)
y2 = np.exp(x) / (1 + np.exp(x))

plt.plot(x, y1)
plt.plot(x, y2)

plt.fill_between(x, y1, y2, alpha = 0.3);
````

````python
x = np.linspace(-np.pi, np.pi, 100)
cos, sin = np.cos(x), np.sin(x)

plt.plot(x, cos)
plt.plot(x, sin)

plt.fill_between(x, cos, sin, where = cos > sin, alpha = 0.3);
````

````python
n = 256
x = np.linspace(-np.pi, np.pi, n, endpoint=True)
y = np.sin(2*x)

plt.plot(x, y+1, color='blue', alpha=1.00)
plt.fill_between(x, 1, y+1, color='blue', alpha = 0.3)

plt.plot(x, y-1, color='blue', alpha=1.00)
plt.fill_between(x, -1, y-1, where = -1 > y-1, color='red', alpha = 0.3)
plt.fill_between(x, -1, y-1, where = -1 < y-1, color='blue', alpha = 0.3)
plt.show()
````

### 1.2.2. Gráficos de dispersión o scatter plots

Un diagrama de dispersión o nube de puntos es un tipo de diagrama matemático que utiliza las coordenadas cartesianas $(x, y)$ de un par de datos, para graficarlos en un plano.

En la descripción se hable de dos dimensiones porque es la idea más intuitiva, y más alla de tres dimensiones no se puede graficar, pero un gráfico de dispersión se puede realizar de cualquier función de n dimensiones.

Graficar una nube de 100 puntos para el espacio lineal $N~(1, 50)$, definida por la función $f(x)=20 + 2x$, donde se le añade un ruido (*jitter*) normal $N~(1, 30)$:

````python
x = np.random.normal(1, 50, size=100)
jitter = np.random.normal(1, 30, size=100)
y = 20 + 2 * x
y_jit = y + jitter

plt.scatter(x, y_jit);
````

Si se quisieran graficar estos puntos mediante una línea, sería necesario ordenar el array para que se graficaran en orden.

````python
plt.plot(x, y_jit);
````

Dibujar una elipse mediante un scatter plot, después intenta rellenarlo:

$$r^2 = sen(\phi)^2 + cos(\phi)^2$$

$$x=r \cdot cos(\phi)$$
$$y=r \cdot sin(\phi)$$

````python
n = 200 # Elevando este número se consigue ir rellenando el círculo 

angle = 2*np.pi*np.random.random(n)
radius = np.sqrt(np.random.random())

x = radius*np.cos(angle)
y = radius*np.sin(angle)

plt.scatter(x, y);
````

````python
plt.plot(x, y);
````

### 1.2.4. Histogramas

Un histograma es una representación gráfica de una variable en forma de barras, donde la superficie de cada barra es proporcional a la frecuencia de los valores representados. Sirven para obtener una primera vista de la distribución de la población, o de la muestra, respecto a una característica, cuantitativa y continua (como la longitud o el peso).

De esta manera ofrece una visión de grupo permitiendo observar una preferencia, o tendencia, por parte de la muestra o población por ubicarse hacia una determinada región de valores dentro del espectro de valores posibles (sean infinitos o no) que pueda adquirir la característica.

Para construir un histograma con `matplotlib` es necesario divir la muestra en intervalos del mismo tamaño, para contabilizar cuantos valores quedan en cada intervalo. Estos intervales se presuponen consecutivos y sin solapamientos.

````python
x_norm = np.random.normal(size = 5000)
y_norm = np.random.normal(size = 25000)

plt.hist(y_norm, bins = 100);
plt.hist(x_norm, bins = 100);
````

````python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=True, facecolor='g', alpha=0.75)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.axis([40, 160, 0, 0.03])
plt.axvline(x.mean(), color = 'orange')
plt.grid(True)
plt.show()
````

## 1.3. Interfaz orientada a objetos

Este método pérmite realizar gráficos muy personalizados y que se ajustan de forma precisa a lo que se quiere mostrar, requieren más trabajo para definir bien todos los elementos.

Mientras que la mayoría de funciones de `pyplot` se pueden traducir de forma directa a los métodos `ax.method()`, (como `plt.plot()` → `ax.plot()`, `plt.legend()` → `ax.legend()`, etc.), existen algunas que no siguen esta norma.

En particular, aquellas funciones para establecer límites, etiquetas y títulos cambian ligeramente cuando se hace mediante objetos. Para realizar la transición entre las funciones estilo MATLAB , y los métodos orientados a objetos, hay que seguir las siguientes relaciones:

- `plt.xlabel()`  → `ax.set_xlabel()`
- `plt.ylabel()` → `ax.set_ylabel()`
- `plt.xlim()`  → `ax.set_xlim()`
- `plt.ylim()` → `ax.set_ylim()`
- `plt.title()` → `ax.set_title()`

También se puede llamar al método `ax.set()` para configurar todas estas propiedades de una sola vez.

### 1.3.1. Subelementos de una figura

**Axes: `figure.add_axes()`**

Método de una figura para añadir un *ax* nuevo a una figura. Permite personalizar el tamaño y posición del *ax* pasando como argumento una lista con los siguientes valores `[izquierda, abajo, ancho, alto]`, los cuales son coordenadas relativas de la figura con valor 0-1.

````python
# Función generadora de puntos aleatorios
def generate_points(t1=2, t2=5):
    x = np.random.normal(1, 50, size=100)
    jitter = np.random.normal(1, 50, size=100)
    y = t1 + t2*x
    y_ = y + jitter
    
    return x, y_
````

````python
# Generación de puntos aleatorios
x, y = generate_points(20, 0.2)

# Generamos un figura de dimensiones 6 x 4
figure = plt.figure(figsize=(6, 4))

# Añadimos un primer ax
ax1 = figure.add_axes([0, 0, 0.8, 0.8])
ax1.scatter(x, y**2)

# Añadimos un segundo ax
ax2 = figure.add_axes([0.75, 0.4, 0.4, 0.3])
ax2.scatter(x, y**2);
````

**Subplots: `figure.add_sublot(nrows, ncolumns, index)`**

Método de una figura para añadir nuevos *axes* a una figura.

````python
x1, y1 = generate_points(50, 1.5)
x2, y2 = generate_points(20, 0.2)

y2_square = y2**2
y2_cube = y2**3

# Generamos una figura de tamaño 10 x 8
figure = plt.figure(figsize=(10,8))

# Generamos un subplot 1 
subplot_1 = figure.add_subplot(1, 2, 1)
subplot_1.scatter(x1, y1)

# Generamos un subplot 2
subplot_2 = figure.add_subplot(2, 2, 2)
subplot_2.scatter(x2, y2_square);

# Generamos un subplot 3
subplot_3 = figure.add_subplot(2, 2, 4)
subplot_3.scatter(x2, y2_cube);
````

**Subplots: `plt.subplots(nrows, ncolumns)`**

Se generan una matriz de dimensiones n x n, donde mediante coordenadas se puede ir configurando los diferenes *axes*. Esta es una de las maneras más cómodas de trabajar con la interfaz orientada a objetos.

````python
# Generar una figura de 10 x 6, con 4 subplots en formato 2x2
figure, axes = plt.subplots(2, 2 ,figsize=(10, 6))
````

````python
# Generamos una figura de 8x6, con 4 subplots en formato 2x2
fig, ax = plt.subplots(2, 2, figsize = (8, 6))

# Definimos una función sigmoide
x = np.linspace(-50, 200, 200)
y = 1/(1+np.exp(-x))

# Las graficamos en cada subplot mediante coordenadas
ax[0,0].plot(x, y);
ax[0,1].scatter(x, y*2);
ax[1,0].plot(x, y**(-1));
ax[1,1].hist(y, bins = 5);
````

**Gridspec: `figure.add_gridspec()`**

Es el método más preciso para generar matrices de subplots. Con este método se puede controlar al detalle el tamaño de cada uno de ellos.

````python
fig = plt.figure(constrained_layout = True)

widths = [2, 3, 1.5]
heights = [1, 3, 2]

spec = fig.add_gridspec(ncols=3, nrows=3, width_ratios=widths,
                          height_ratios=heights)
for row in range(3):
    for col in range(3):
        ax = fig.add_subplot(spec[row, col])
        label = 'Width: {}\nHeight: {}'.format(widths[col], heights[row])
        ax.annotate(label, (0.1, 0.5), xycoords='axes fraction', va='center')
````

### 1.3.3. Ejes compartidos

Mediante los argumentos `sharex` y `sharey` se pueden igualar las escalas de los ejes $x$ e $y$ de una matriz de subplots.

````python
x = np.linspace(-2, 2, 100)

fig, ax = plt.subplots(2, 2, sharex = True, sharey = True)

ax[0,0].plot(x, 2*x);
ax[0,1].plot(x, np.abs(x));
ax[1,0].plot(x, x**2);
ax[1,1].plot(x, np.sin(x));
````

### 1.3.6. Parametrización de los elementos de un gráfico

Existen multitud de parámetros de las líneas que pueden ser modificados para personalizar los gráficos de `matplotlib`.

**Color:**

Se puede definir el color de una línea de muchas formas:

````python
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x - 0), color='blue')           # specify color by name
plt.plot(x, np.sin(x - 1), color='g')              # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')           # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')        # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3))  # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse');    # all HTML color names supported
````

También se pueden colorear las líneas mediante los mapas de color propios de la librería `matplotlib`. Los mapas de color son conjuntos de colores que transicionan a lo largo de una gama, con el objetivo de resaltar de una cierta manera un gráfico.

En el siguiente enlace se puede leer acerca de los mismos: https://matplotlib.org/users/colormaps.html

# 2. Seaborn

`seaborn` es una biblioteca de visualización de datos de Python basada en `matplotlib`. Proporciona una interfaz de alto nivel para dibujar gráficos estadísticos atractivos e informativos. Al igual que `matplotlib` se basa en arrays de `numpy` par trabajar, **`seaborn` trabaja de forma nativa con `pandas`**.

Aunque no es tan completa o personalizable como `matplotlib`, es más sencilla y el código necesario para representar lo mismo es más corto.

## 2.1. Líneas

````python
x = np.linspace(-5, 5, 100)
y1 = np.cos(x)
y2 = np.exp(x) / (1 + np.exp(x))

data = {'x': x, 'y1': y1, 'y2': y2}
df = pd.DataFrame(data)

sns.lineplot(data=df, x='x', y='y1', label='y1')
sns.lineplot(data=df, x='x', y='y2', label='y2')

plt.legend(loc='upper left');
````


````python
sns.relplot(data=df, x='sepal_length', y='sepal_width');
````

````python
sns.relplot(data=df, x='sepal_length', y='sepal_width', style='species', hue='species');
````

## 2.7. Variables categóricas

Hasta ahora la mayoría de los gráficos que se han visto tenían como objetivo representar variables continuas, o discretas pero con multitud de posibles valores, en los siguientes puntos se analizarán diferentes métodos representativos para analizar variables categóricas.

Como veremos más adelante, el método `sns.catplot()` permite indicar mediante el parámetro `kind` el tipo de gráfico. Para cada tipo existe adicionalmente un método propio de `seaborn`, la relación es la siguiente:

- `sns.catplot(kind='swarm')` -> `sns.swarmplot()`
- `sns.catplot(kind='strip')` -> `sns.stripplot()`
- `sns.catplot(kind='box')` -> `sns.boxplot()`
- `sns.catplot(kind='violin')` -> `sns.violinplot()`

El método propio de seaborn permite colocar los gráficos de salida en *axes* de `matplotlib`.

### 2.7.1. Diagramas de dispersión catgóricos: swarm (enjambre) y strip (tira) plot

En realidad, hay dos diagramas de dispersión categóricos diferentes en `seaborn`. Adoptan diferentes enfoques para resolver el desafío principal de representar datos categóricos con un diagrama de dispersión, que es que todos los puntos pertenecientes a una categoría caerían en la misma posición a lo largo del eje correspondiente a la variable categórica. Lo que cambia entre un modo u otro, es como se agrupan estos puntos respecto a un eje.

Estos gráficos permiten obtener una idea visual de cómo se distribuye una variable cotinua asociada a otra categórica.

```Python
sns.catplot(

    kind      # Modifica si el plot ('strip', 'swarm')

```

### 2.7.2. Distribución de observaciones categóricas: boxplots y violionplots

A medida que aumenta el tamaño del conjunto de datos, los diagramas de dispersión categóricos se limitan en la información que pueden proporcionar sobre la distribución de valores dentro de cada categoría. Cuando esto sucede, existen varios enfoques para resumir la información distributiva de manera que faciliten las comparaciones entre variables.

El Diagrama de Caja o Boxplot es un tipo de gráfico que muestra el resumen de una variable mediante cinco medidas descriptivas, además de intuir su morfología y simetría.

Este tipo de gráficos permite identificar valores atípicos y comparar distribuciones. Además de conocer de una forma cómoda y rápida como se distribuyen el 50% de los valores centrales.

- Caja: indica donde se encuentra la gran mayoría de los datos, el borde inferior de la caja muestra el percentil 25, y el superior el percentil 75; por lo que se puede inferir que la caja muestra el 50% de los datos. La raya amarilla indica el percentil 50, o la mediana. Por lo que la caja, me sirve para interpretar dónde se condensan los valores centrales. Si la mediana se sitúa en el centro de la caja entonces la distribución es simétrica y tanto la media, mediana y moda coinciden; si la mediana corta la caja en dos partes desiguales, la media será diferente a la mediana.

- Bigote: Habrá un límite superior, que no podrá superar el 1,5  veces el IQR, si el máximo no supera ese valor, la longitud del bigote será desde el tercer cuartil hasta el máximo. Habrá un límite inferior, que no podrá superar el 1,5  veces el IQR, si el mínimo no supera ese valor, la longitud del bigote será desde el primer cuartil hasta el mínimo.

- Outliers: Son aquellos valores que quedan por encima o debajo del máximo del bigote, es decir 1,5*IQR. Se reresentan con bolitas.
